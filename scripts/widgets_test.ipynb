```python
import itkwidgets
import tomobase
from tomobase import phantoms
import tomobase.processes.forward_project
import tomobase.tiltschemes
import plotly.express as px
scheme = tomobase.tiltschemes.GRS(-70,70, 0)
angles = [scheme.get_angle() for i in range(60)]
sino = tomobase.processes.project(phantoms.nanocage(),angles)
#itkwidgets.view(sino.data)
img = sino.data[0,:,:]
fig = px.imshow(img)
fig.show()
```

```python
import numpy as np
from PIL import Image
import ipywidgets as widgets
import io
from IPython.display import display

# Replace this with your own data slice
arr = sino.data[0, :, :]  # assuming float32 or float64

# Normalize and convert to uint8
arr_min, arr_max = arr.min(), arr.max()
arr_scaled = ((arr - arr_min) / (arr_max - arr_min + 1e-8) * 255).astype(np.uint8)

# Convert to a displayable grayscale image
img = Image.fromarray(arr_scaled, mode='L')

# Encode to PNG in memory
with io.BytesIO() as output:
    img.save(output, format='PNG')
    image_bytes = output.getvalue()

# Display as ipywidget
img_widget = widgets.Image(value=image_bytes, format='png')
display(img_widget)
```
