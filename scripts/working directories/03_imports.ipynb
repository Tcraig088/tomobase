```python
import tomobase
import os
import tomobase.tiltschemes
import numpy as np

ts = tomobase.tiltschemes.Incremental(-64, 64, 2)
angles = np.array([ts.get_angle() for i in range(65)])
folder = r'\\ematbyname\emat\TimC\misc\Simulations\SIM_SB_TC_004\vol'
folder_fixed = r'\\ematbyname\emat\TimC\misc\Simulations\SIM_SB_TC_002\vol'
file_fixed = '0_vol_init.rec'
sino = None
i=0
j=0
k=0
index = 3
_list = [None]*80
for file in os.listdir(folder):
    if file.endswith('.rec'):
        if '00' not in file:
            _list[i] = file
            i+=1

print(_list)
for i, angle in enumerate(angles):
    path = os.path.join(folder, _list[j])
    vol = tomobase.data.Volume.from_file(path)
    print(path)
    sino2 = tomobase.processes.project(vol, [angles[i]])
    if sino is None:
        sino = sino2
    else:
        sino.insert(sino2.data[:,:,0], angles[i])
    k+=1
    if k == index:
        j+=1
        k=0


print(angles[i])
sino.to_file('inc_blurred.mrc')
rec = tomobase.processes.reconstruct_weighted_sirt(sino, iterations=100, weighted = True)
rec.to_file('inc_blurred.rec')
```
